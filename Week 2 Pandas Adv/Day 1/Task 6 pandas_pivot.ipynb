<h2 style="color:brown">Grouper</h2>

```python
df = pd.read_csv(r"D:\Saad Maqbool\New folder\weather3.csv")
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df
```

```python
# Pivot table with monthly grouping
df.pivot_table(index=pd.Grouper(freq='ME', key='date'), columns='city')
```

```python
df = df.sort_values('date')

df['temp_rolling_3'] = df['temperature'].rolling(window=3).mean()
df['humidity_rolling_3'] = df['humidity'].rolling(window=3).mean()

df['temp_ema'] = df['temperature'].ewm(span=3).mean()

print("Rolling Averages:")
print(df[['date', 'temperature', 'temp_rolling_3', 'temp_ema']])
```

```python
pivot_complex = df.pivot_table(
    index=pd.Grouper(key='date', freq='M'),
    values=['temperature', 'humidity'],
    aggfunc=['mean', 'max', 'min', 'std', 'count']
).round(2)

print("Monthly Pivot with Multiple Aggregations:")
print(pivot_complex)
```
