```python
def daily_summary_multi(df, n_hours_per_day=24):
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    n_rows = df.shape[0]

    if n_rows % n_hours_per_day != 0:
        raise ValueError(f"Row count ({n_rows}) is not divisible by {n_hours_per_day}")

    n_days = n_rows // n_hours_per_day
    summaries = []

    for day in range(n_days):
        start = day * n_hours_per_day
        end = start + n_hours_per_day
        day_df = df.iloc[start:end]

        day_summary = {"day": day + 1}

        for col in numeric_cols:
            values = day_df[col].to_numpy()
            day_summary[f"{col}_min"] = round(float(np.nanmin(values)), 2)
            day_summary[f"{col}_max"] = round(float(np.nanmax(values)), 2)
            day_summary[f"{col}_mean"] = round(float(np.nanmean(values)), 2)

        summaries.append(day_summary)

    return summaries


summaries = daily_summary_multi(df)
#for 2 value print 
for s in summaries[:2]:
    print(s)
```

```python
def daily_summary_multi(df, column_rename, n_hours_per_day=24):
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    n_rows = df.shape[0]
    if n_rows % n_hours_per_day != 0:
        raise ValueError(f"Row count ({n_rows}) is not divisible by {n_hours_per_day}.")
    n_days = n_rows // n_hours_per_day
    summaries = []
    for day in range(n_days):
        start = day * n_hours_per_day
        end = start + n_hours_per_day
        day_df = df.iloc[start:end]
        day_summary = {"day": day + 1}
        for col in numeric_cols:
            short_name = column_rename.get(col, col)
            values = day_df[col].to_numpy()
            day_summary[f"{short_name}_min"] = round(float(np.nanmin(values)), 2)
            day_summary[f"{short_name}_max"] = round(float(np.nanmax(values)), 2)
            day_summary[f"{short_name}_mean"] = round(float(np.nanmean(values)), 2)
        summaries.append(day_summary)
    return summaries
```
